## Функция парсинга данных из.csv / скачивание pdf / сохранение в .md и IMAGES

```python
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
import fitz  # !pip install PyMuPDF
from PIL import Image
from io import BytesIO

def pars_csv(path_csv: str, path_pdf: str, path_text: str, path_image: str, start_row_num:int, path_settings:str, path__account_json:str, read:bool): # start_row_num - строка, с которой начинаем обрабатывать данные / read:bool - True / False
    # start_row_num = 3  # Параметр для указания номера строки, с которой начинать обработку данных
    start_row_num = start_row_num - 1  # Переводим в систему исчисления, начиная с "0"

    # Чтение скачанного файла
    data = pd.read_csv(path_csv, skiprows=range(1, start_row_num))

    # Проверка наличия столбца "ID" и прекращение считывания строк, если он отсутствует
    if "ID" not in data.columns:
        status_op = False
        print("Столбец 'ID' отсутствует")
    else:
        # Находим заголовок "НАЗВАНИЕ МОДЕЛИ" в строке с заголовками
        column_name = None
        for col in data.columns:
            if "НАЗВАНИЕ МОДЕЛИ" in col:
                column_name = col
                break

        if column_name is None:
            status_op = False
            print("Столбец 'НАЗВАНИЕ МОДЕЛИ' не найден")
        else:
            status_op = True
            # Создание файлов и запись данных
            for index, row in data.iterrows():
                if pd.notna(row['ID']):  # Проверка наличия ID в текущей строке
                    model_name = str(row[column_name])
                    file_name = os.path.join(path_text, model_name + ".md")

                    # Скачивание файла PDF
                    pdf_url = row['ССЫЛКА НА ФАЙЛ С ОПИСАНИЕМ']
                    if pd.notna(pdf_url):
                        pdf_path = os.path.join(path_pdf, model_name + ".pdf")
                        response = requests.get(pdf_url)
                        if response.status_code == 200:
                            with open(pdf_path, 'wb') as pdf_file:
                                pdf_file.write(response.content)
                        print(f"Загружен файл '{model_name}.pdf'") # Отображение имени загружаемого файла
                        
                        # Извлечение текста из файла PDF
                        text_content = ''
                        with fitz.open(pdf_path) as pdf_file:
                            for page_num in range(pdf_file.page_count):
                                page = pdf_file[page_num]
                                text_content += page.get_text("text")
                            
                        # Запись информации в файл .md
                        with open(file_name, "w") as file:
                            file.write(f"# {model_name}\n\n")
                            file.write(f"Текстовое содержание:\n{text_content}\n\n")
                            file.write(f"# ПРИМЕРНАЯ ЦЕНА\n{row['ПРИМЕРНАЯ ЦЕНА']}\n\n")
                            file.write(f"# ДАТА АКТУАЛИЗАЦИИ\n{row['ДАТА АКТУАЛИЗАЦИИ']}\n")

                        # Сохранение изображения
                        image_url = row['ИЗОБРАЖЕНИЕ']
                        response = requests.get(image_url)
                        if response.status_code == 200:
                            img = Image.open(BytesIO(response.content))
                            image_path = os.path.join(path_image, model_name + '.jpg')
                            img.save(image_path)
                        else:
                            status_op = False  # Ссылка из столбца "ССЫЛКА НА ФАЙЛ С ОПИСАНИЕМ" не прочиталась

                    row_id = str(row['ID'])
                    print("\n=======\nID строки:", row_id)
                    status_op = str(status_op)
                    print("status_op:", status_op)
                    write_data_to_google_sheet(path_settings, row_id, path__account_json, read)
                    
                else:
                    print("ID отсутствует, прекращаем обработку строк")
                    break  # Прекращаем обработку строк, если ID отсутствует
        
    print("Считывание завершено")

path_csv = "PROJECT/DATA/GOOGLE/IN/LOAD_GOOGLE_CSV/google.csv"
path_pdf = 'PROJECT/DATA/DOC_LOADED'
path_text = 'PROJECT/DATA/TEXT_MD'
path_image = 'PROJECT/DATA/IMAGES'
start_row_num = 3
path_settings = "PROJECT/SETTINGS/settings.yml"
path__account_json = 'PROJECT/DATA/GOOGLE/test-neuro-sotrudnik-9d66b2f82586.json'
read = "True" # True / False

pars_csv(path_csv, path_pdf, path_text, path_image, start_row_num, path_settings, path__account_json, read)
```

## Текст md в чанки md

```python
# Делим на чанки
# import os
# import shutil

def process_md_files(input_folder, output_folder):
    # Вспомогательные функции
    # Очистка конечной папки перед обработкой
    clear_folder(output_folder)
    # Чтение текстового md-файла
    def read_md_files(input_folder): # (вспомогательная)
        files_content = {}
        for filename in os.listdir(input_folder):
            if filename.endswith(".md"):
                with open(os.path.join(input_folder, filename), 'r', encoding='utf-8') as file:
                    files_content[filename] = file.read()
        return files_content
    # Деление на чанки
    def split_into_chunks(text, max_chunk_size=500):
        lines = text.split('\n')
        chunks = []
        current_chunk = []
        current_chunk_size = 0

        def start_new_chunk():
            nonlocal current_chunk, current_chunk_size
            if current_chunk:
                chunks.append('\n'.join(current_chunk))
            current_chunk = ['## Чанк']
            current_chunk_size = len('## Чанк')
        start_new_chunk()
        
        for line in lines:
            if line.startswith('# ') or line.startswith('## '):
                if current_chunk_size + len(line) > max_chunk_size:
                    start_new_chunk()
            
            current_chunk.append(line)
            current_chunk_size += len(line)

            if current_chunk_size >= max_chunk_size:
                start_new_chunk()
        
        if current_chunk:
            chunks.append('\n'.join(current_chunk))
        
        return chunks
    # Сохранение чанков в текстовый md с обозначенными чанками
    def save_chunks(output_folder, chunks, filename):
        output_path = os.path.join(output_folder, filename)
        with open(output_path, 'w', encoding='utf-8') as file:
            for chunk in chunks:
                file.write(chunk + '\n\n')

    # RUN
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    files_content = read_md_files(input_folder)
    for filename, text in files_content.items():
        chunks = split_into_chunks(text)
        # print(chunks)
        save_chunks(output_folder, chunks, filename)
    print("Файл с разделенным текстом на чанки сохранен")

input_folder = 'PROJECT/DATA/TEXT_MD'
output_folder = 'PROJECT/DATA/SHUNKS_MD'
# Обработка файлов
process_md_files(input_folder, output_folder)
```

## Чанки MD в FAISS

```python
# ПОДГГОТОВКА БАЗЫ из MD в FAISS
import os
import faiss
import numpy as np
import pickle
from langchain.embeddings import OpenAIEmbeddings


# Директории
def process_files(chunks_dir: str, db_dir: str):
    # Очищаем папку
    clear_folder(db_dir)

    # Вспомогательные функции
    def load_text_from_file(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            return file.read()

    def split_text_into_chunks(text):
        chunks = text.split('## Чанк')
        return [chunk.strip() for chunk in chunks if chunk.strip()]

    def create_faiss_index(chunks, embeddings):
        # Получение векторных представлений
        vectors = embeddings.embed_documents(chunks)

        # Преобразование в numpy массив, если необходимо
        if isinstance(vectors, list):
            vectors_np = np.array(vectors).astype('float32')
        else:
            vectors_np = np.array(vectors).astype('float32')

        # Проверка правильности формы массива
        if vectors_np.ndim != 2:
            raise ValueError("Vectors should be a 2D numpy array.")

        # Создание FAISS индекса
        d = vectors_np.shape[1]
        index = faiss.IndexFlatL2(d)
        index.add(vectors_np)
        return index, vectors_np

    def save_faiss_index_and_chunks(index, vectors_np, chunks, file_base):
        # Сохранение индекса
        index_file_path = os.path.join(db_dir, file_base + '.faiss')
        faiss.write_index(index, index_file_path)

        # Сохранение векторов и текстов чанков
        data_file_path = os.path.join(db_dir, file_base + '_data.pkl')
        with open(data_file_path, 'wb') as f:
            pickle.dump({'vectors': vectors_np, 'chunks': chunks}, f)

        print(f'Индекс FAISS и данные для {file_base} сохранены.')

    files = [f for f in os.listdir(chunks_dir) if os.path.isfile(os.path.join(chunks_dir, f))]

    embeddings = OpenAIEmbeddings()  # Инициализация модели для эмбеддингов

    for file_name in files:
        file_path = os.path.join(chunks_dir, file_name)
        text = load_text_from_file(file_path)
        chunks = split_text_into_chunks(text)
        index, vectors_np = create_faiss_index(chunks, embeddings)
        file_base = os.path.splitext(file_name)[0]
        save_faiss_index_and_chunks(index, vectors_np, chunks, file_base)

chunks_dir = 'PROJECT/DATA/SHUNKS_MD'
db_dir = 'PROJECT/DATA/DB_FAISS'
process_files(chunks_dir, db_dir)
```

```python
# Функция изменения статуса в Google
def write_data_to_google_sheet(path__account_json:str, path_settings:str,id_str:str, name_vkladka:str, read="True"): # Функция изменения статуса в Google
    gc = gspread.service_account(filename=path__account_json)
    
    # Получаем значение ключей
    url_data_links_ = get_environment_value(path_settings, "url_data_links")
    print("\nЗагруженная ссылка на таблицу: ", url_data_links_) 
    name_vkladka_ = get_environment_value(path_settings, name_vkladka) # "name_vkladka_olther"
    print("\nЗагруженное название вкладки: ", name_vkladka_) 
    # name_vkladka_contacts_ = get_environment_value(file_path, "name_vkladka_contacts")
    # print("\nЗагруженное название вкладки: ", name_vkladka_contacts_) 

    # URL таблицы Google Sheets
    sheet_url = url_data_links_ #'URL_ВАШЕЙ_ТАБЛИЦЫ'
    
    # Получение ID таблицы из URL
    # sheet_id = get_sheet_id_from_url(sheet_url)
    def get_sheet_id_from_url(url):
        # Регулярное выражение для извлечения ID таблицы из URL
        pattern = r'/d/([a-zA-Z0-9-_]+)'
        match = re.search(pattern, url)
        if match:
            return match.group(1)
        else:
            raise ValueError("Неверный URL-адрес Google Таблиц»")
    # Получение ID таблицы из URL
    sheet_id = get_sheet_id_from_url(sheet_url)
    print("ID таблицы: ", sheet_id)

    # Открытие таблицы по ID
    sh = gc.open_by_key(sheet_id)
    
    # Вносим данные во вкладку "OLTHER"
    # Название вкладки
    sheet_name = name_vkladka_ #'НАЗВАНИЕ_ВКЛАДКИ'
    
    # Получение данных из указанной вкладки
    worksheet = sh.worksheet(sheet_name)
    data = worksheet.get_all_values()
    
    # Номер строки с заголовками столбцов
    str_header = 1
    str_header = str_header -1
    
    # Находим индекс столбца с заголовком "ID"
    index_id = data[str_header].index('ID')
    
    # Значение, по которому будем искать нужную строку
    # id_str = '2'
    
    # Находим индекс строки, где значение в колонке "ID" равно id_str
    for i, row in enumerate(data):
        if i > str_header and row[index_id] == id_str:
            target_row_index = i
            break
    
    # Название столбца, в который будем записывать данные
    title_status = 'СТАТУС'
    
    # Находим индекс столбца с заголовком "СТАТУС"
    index_status = data[str_header].index(title_status)
    
    # Записываем значение в ячейку строки с найденным ID и указанного столбца
    if read == "True":
        worksheet.update_cell(target_row_index + 1, index_status + 1, 'Внесено в базу')
    elif read == "False":
        worksheet.update_cell(target_row_index + 1, index_status + 1, 'Битая ссылка')
    else:
        print("Error data")
    
    print(f"Данные успешно записаны в Google таблицу '{sheet_name}'")

# id_str = "1"
# read = "True" # True / False
# path__account_json = 'PROJECT/DATA/GOOGLE/test-neuro-sotrudnik-9d66b2f82586.json'
# path_settings = "PROJECT/SETTINGS/settings.yml"
# name_vkladka = "name_vkladka_olther"
# write_data_to_google_sheet(path__account_json, path_settings, id_str, name_vkladka)
```

### GOOGLE_OLTHER_CSV_TO_MD

```python
# GOOGLE_OLTHER_MD

import os
import csv

# Функция для чтения данных из CSV файла
def process_csv_to_md_other(path_google_olther_csv:str, str_heder:int, str_start:int, path_md_google_olther:str, path__account_json:str, path_settings:str):
    clear_folder(path_md_google_olther)
    # Проверяем наличие и создаем директорию, если её нет
    # data_folder = "3_DATA/DATA_MD_GOOGLE_OLTHER"
    
    # Создаем все промежуточные директории, если их нет
    try:
        os.makedirs(path_md_google_olther, exist_ok=True)
    except OSError as e:
        print(f"Ошибка при создании директории {path_md_google_olther}: {e}")
        return
    
    # Проверяем наличие файла, и если его нет, то создаем
    output_file = os.path.join(path_md_google_olther, "data_google_olther.md")
    if not os.path.exists(output_file):
        with open(output_file, 'w'): pass  # Создаем пустой файл
    
    # Открываем CSV файл для чтения
    with open(path_google_olther_csv, mode='r', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        
        # Пропускаем строки до строки заголовков
        for _ in range(str_heder - 1):
            next(csv_reader)
        
        headers = next(csv_reader)  # Читаем заголовки
        
        # Находим индексы нужных колонок
        try:
            id_index = headers.index("ID")
            tema_index = headers.index("ТЕМА")
            info_index = headers.index("ИНФОРМАЦИЯ")
        except ValueError:
            print("Одна из требуемых колонок отсутствует в файле CSV.")
            return
        
        # Читаем строки и обрабатываем данные
        for idx, row in enumerate(csv_reader, start=1):
            if idx >= str_start - 1:
                id_value = row[id_index].strip() if id_index < len(row) else ""
                id_value = str(id_value)
                tema_value = row[tema_index].strip() if tema_index < len(row) else ""
                info_value = row[info_index].strip() if info_index < len(row) else ""
                
                if tema_value and info_value:
                    # Формируем данные для записи в файл Markdown
                    with open(output_file, 'a', encoding='utf-8') as md_file:
                        # Записываем заголовок и информацию
                        md_file.write(f"# {tema_value}\n")
                        md_file.write(f"{info_value}\n\n")
                # Вносим СТАТУС строки
                name_vkladka = "name_vkladka_olther"
                write_data_to_google_sheet(path__account_json, path_settings, id_value, name_vkladka)


# Запускаем обработку данных из CSV файла
path_google_olther_csv = "PROJECT/DATA/GOOGLE/IN/LOAD_OLTHER_GOOGLE_CSV/google_olther.csv"
str_heder = 1  # Номер строки заголовков
str_start = 3  # Номер строки, с которой начинаем чтение данных
path_md_google_olther="PROJECT/DATA/DATA_MD_GOOGLE_OLTHER"
path__account_json = 'PROJECT/DATA/GOOGLE/test-neuro-sotrudnik-9d66b2f82586.json'
path_settings = "PROJECT/SETTINGS/settings.yml"

process_csv_to_md_other(path_google_olther_csv, str_heder, str_start, path_md_google_olther, path__account_json, path_settings)
```

### MD_CHUNK_OLTHER_CONTACTS_FAISS

```python
# ПОДГГОТОВКА БАЗЫ из MD в FAISS
import os
import faiss
import numpy as np
import pickle
from langchain.embeddings import OpenAIEmbeddings


def process_files(path_chunks_md:str, path_db_faiss:str):
    # Создание директории для хранения векторных баз, если не существует
    if not os.path.exists(path_db_faiss):
        os.makedirs(path_db_faiss)

    # Загрузка текстов из файлов
    def load_text_from_file(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            return file.read()

    # Разделение текста на чанки
    def split_text_into_chunks(text):
        # Простое разделение текста по строкам
        chunks = text.split('# ')
        return [chunk.strip() for chunk in chunks if chunk.strip()]

    # Создание векторного хранилища
    def create_faiss_index(chunks, embeddings):
        # Использование embed_documents для получения векторных представлений
        vectors = embeddings.embed_documents(chunks)
        vectors_np = np.array(vectors).astype('float32')
        
        # Создание FAISS индекс
        index = faiss.IndexFlatL2(vectors_np.shape[1])
        index.add(vectors_np)
        return index, vectors_np

    # Сохранение FAISS индекса и текстов чанков
    def save_faiss_index_and_chunks(index, vectors_np, chunks, file_base):
        # Сохранение индекса
        index_file_path = os.path.join(path_db_faiss, file_base + '.faiss')
        faiss.write_index(index, index_file_path)
        
        # Сохранение векторов и текстов чанков в формате pkl
        data_file_path = os.path.join(path_db_faiss, file_base + '_data.pkl')
        with open(data_file_path, 'wb') as f:
            pickle.dump({'vectors': vectors_np, 'chunks': chunks}, f)
        
        print(f'FAISS index and data for {file_base} saved.')


    files = [f for f in os.listdir(path_chunks_md) if os.path.isfile(os.path.join(path_chunks_md, f))]
    embeddings = OpenAIEmbeddings()  # Инициализация модели для эмбеддингов
    for file_name in files:
        file_path = os.path.join(path_chunks_md, file_name)
        text = load_text_from_file(file_path)
        
        # Разделение текста на чанки
        chunks = split_text_into_chunks(text)
        
        # Создание и сохранение FAISS индекса и текстов чанков
        index, vectors_np = create_faiss_index(chunks, embeddings)
        file_base = os.path.splitext(file_name)[0]
        save_faiss_index_and_chunks(index, vectors_np, chunks, file_base)

print('База Контакты')
path_chunks_md = 'PROJECT/DATA/DATA_MD_GOOGLE_CONTACTS'
path_db_faiss = 'PROJECT/DATA/DB_FAISS_CONTACTS'
process_files(path_chunks_md, path_db_faiss)

print('База Прочее')
path_chunks_md = 'PROJECT/DATA/DATA_MD_GOOGLE_OLTHER'
path_db_faiss = 'PROJECT/DATA/DB_FAISS_OLTHER'
process_files(path_chunks_md, path_db_faiss)
```
